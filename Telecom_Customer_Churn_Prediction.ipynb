#### Просмотр пропущенных значений в данных

```python
print("Сумма пропусков в train:", train.isna().sum().sum())
print("Сумма пропусков в test:", test.isna().sum().sum())
```

#### Замена пропусков средними значениями в "численных" колонках 

```python
train_numeric = train.select_dtypes(include=np.number)
train_numeric_columns = train_numeric.columns
for i in train_numeric_columns:
    train[i] = train[i].fillna(train[i].mean())
print("Сумма пропусков в train:", train.isna().sum().sum())

test_numeric = test.select_dtypes(include=np.number)
test_numeric_columns = test_numeric.columns
for i in test_numeric_columns:
    test[i] = test[i].fillna(test[i].mean())
print("Сумма пропусков в test:", test.isna().sum().sum())
```

```python
customer_id_train = train['Customer ID']
train= train.drop(columns='Customer ID', axis=1)
```

```python
customer_id_test = test['Customer ID']
test= test.drop(columns='Customer ID', axis=1)
```

#### Замена пропусков наиболее чаще встречающимся значением в категориальных столбцах и просмотр размера данных

```python
train_categorical = train.select_dtypes(include=object)
train_categorical_columns = train_categorical.columns
for i in train_categorical_columns:
    train[i] = train[i].fillna(train[i].mode()[0])
print("Сумма пропусков в train:", train.isna().sum().sum(), "Размер train:", train.shape, sep='\n')

test_categorical = test.select_dtypes(include=object)
test_categorical_columns = test_categorical.columns
for i in test_categorical_columns:
    test[i] = test[i].fillna(test[i].mode()[0])
print("Сумма пропусков в test:", test.isna().sum().sum(), "Размер test:", test.shape, sep='\n')    
```

#### Нормализация

```python
for col in train_categorical_columns:
    uniq_train = pd.unique(train[col])
    replace_dict_train = dict(zip(uniq_train, range(1, len(uniq_train) + 1)))
    train[col] = train[col].replace(replace_dict_train)

for col in test_categorical_columns:
    uniq_test = pd.unique(test[col])
    replace_dict_test = dict(zip(uniq_test, range(1, len(uniq_test) + 1)))
    test[col] = test[col].replace(replace_dict_test)    
```

```python
for col in train.columns:
    train[col] = (train[col] - min(train[col])) / (max(train[col]) - min(train[col]))
train.head()
```

#### Присоединение колонки айдишника так как она нам еще понадобится для сабмита на кэггл

```python
customer_id_train = customer_id_train.to_frame()
train = customer_id_train.join(train, how='right')
train.head()
```

```python
for col in test.columns:
    test[col] = (test[col] - min(test[col])) / (max(test[col]) - min(test[col]))
test.head()
```

```python
customer_id_test = customer_id_test.to_frame()
test = customer_id_test.join(test, how='right')
test.head()
```

#### Смотрю количество уходов/не уходов и балансирую классы с помощью семплирования, для более точного предсказания 

```python
f, ax = plt.subplots(figsize = (7, 5))
sns.countplot(x = 'Customer Status', data = train)
```

```python
train_target_count_0, train_target_count_1 = train['Customer Status'].value_counts()
train_target_0 = train[train['Customer Status'] == 0]
train_target_1 = train[train['Customer Status'] == 1]
print('train target 0:', train_target_0.shape)
print('train target 1:', train_target_1.shape)
#Балансирую классы
train_target_1_over = train_target_1.sample(train_target_count_0, replace = True)
train_over = pd.concat([train_target_1_over, train_target_0], axis = 0)
print("total train target of 1 and 0:\n", train_over['Customer Status'].value_counts())
train_over['Customer Status'].value_counts().plot(kind='bar', title='count (train target)')
train_over_shuffled = train_over.sample(frac=1)
```

#### Дропаю таргет для последущего обучения модели 

```python
X_train = train_over_shuffled.drop(columns='Customer Status', axis=1)
y_train = train_over_shuffled['Customer Status'].values
X_test = test
```

#### Дропаю колонки, чтобы был одинаковый размер для обучения модели и удаля айдишник, тк он не нужен при обучении

```python
X_train = X_train.drop(columns = ['Customer ID', 'Churn Category','Churn Reason'],axis = 1)
X_test = X_test.drop(columns = ['Customer ID'],axis = 1)
```

#### Импорт и обучение 

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10)
rf.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(max_depth=2, n_estimators=50, random_state=12, learning_rate=1)
gb.fit(X_train, y_train)
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.svm import SVC

estimators = [('lr', LogisticRegression()), ('dt', DecisionTreeClassifier())]
stClf = StackingClassifier(estimators=estimators, final_estimator=SVC())
stClf.fit(X_train, y_train)
```

```python
from sklearn.ensemble import BaggingClassifier
bgClf = BaggingClassifier(base_estimator=LogisticRegression(), n_estimators=10, random_state=12)
bgClf.fit(X_train, y_train)
```

```python
from sklearn.ensemble import AdaBoostClassifier

adaClf = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=2), n_estimators=50, random_state=12)
adaClf.fit(X_train, y_train)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

model = Sequential()
model.add(Dense(128, input_shape=(X_train.shape[1],), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(16, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(8, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_val, y_val))
```

#### Сабмит на кэггл

```python
customer_id_test1 = test['Customer ID']
customer_id_test1
```

```python
predict_prob=model.predict(X_test)

classes_prob=np.argmax(predict_prob,axis=1)
```

Таким образом, я обучила несколько моделей, лучший результат показал Bagging Classifier с эстиматором LogisticRegression(0.73089).\
Логистическая регрессия, дерево решений, случайный лес и градиентный бустинг - классические модели мл, но выбрала я их не просто так.\
Логистическая регрессия хорошо работает,когда данные линейно разделимы. Второе место после Bagging Classifier.\
Дерево решений может обрабатывать нелинейные взаимосвязи и взаимодействия между признаками, но склонна к переобучению при большой глубине дерева.
Случайный лес уменьшает переобучение по сравнению с деревом решений, может рабоать с нелинейными взаимосвязями.\
Градиентный бустинг хорошо обучается на сложных данных, может обрабатывать нелинейные взаимосвязи, использует слабые модели для создания сильной модели, но требует тщательной настройки гиперпараметров.\
Также попробовала ансамблевые методы. Они снижают переобучение, устойчивы к выбросам, могут обрабатывать сложные взаимосвязи между фичами, уменьшают смещение и дисперсию, легкие в использовании.\
Нейронные сети должны были показать выше скор, я ожидала их видеть на первом месте, но скорее всего я сделала что-то не так и модель переобучилась. Нейронки могут эффективно моделировать сложные нелинейные зависимости между признаками, также могут автоматически извлекать важные признаки из данных и это еще далеко не все хорошие их качества:D\
Я выбрала оптимайзер Adam, потому что он автоматически адаптирует скорость обучения для каждого параметра в процессе обучения. Ну и он часто хорошо себя показывает на практике в сравнении с другими оптимизаторами.\
Я не выбрала такие алгоритмы мл как линейныя регресиия и knn, потому что линейная регрессия слишком тривиальная модель, она не может улавливать сложные нелинейные связи в данных. Knn чувствительна к шумам и выбросам. В пространствах высокой размерности все точки становятся "далекими" друг от друга, и KNN может терять свою эффективность, поскольку становится сложно определить ближайших соседей, поэтому он медленный и неэффективный для больших объемов данных.
